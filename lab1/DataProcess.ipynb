```python
# Создаем словарь сокращений
short_names = {
    'Location': 'Loc',
    'MinTemp': 'MinT', 
    'MaxTemp': 'MaxT',
    'Rainfall': 'Rain',
    'Evaporation': 'Evap',
    'Sunshine': 'Sun',
    'WindGustDir': 'WGDir',
    'WindGustSpeed': 'WGSpeed',
    'WindDir9am': 'WD9am',
    'WindDir3pm': 'WD3pm',
    'WindSpeed9am': 'WS9am',
    'WindSpeed3pm': 'WS3pm',
    'Humidity9am': 'H9am',
    'Humidity3pm': 'H3pm',
    'Pressure9am': 'P9am',
    'Pressure3pm': 'P3pm',
    'Cloud9am': 'C9am',
    'Cloud3pm': 'C3pm',
    'Temp9am': 'T9am',
    'Temp3pm': 'T3pm',
    'RainToday': 'RainToday'
}

df.rename(columns=short_names, inplace=True)

# Автоматически находим числовые столбцы
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f"Числовые признаки: {numeric_columns}")
print(f"Количество числовых признаков: {len(numeric_columns)}")

# Проверим наличие столбца RainTomorrow и его значения
if 'RainTomorrow' in df.columns:
    print(f"\nРаспределение RainTomorrow:")
    print(df['RainTomorrow'].value_counts())
    print(f"Тип данных RainTomorrow: {df['RainTomorrow'].dtype}")
else:
    print("Столбец RainTomorrow не найден в данных!")
    # Если нет RainTomorrow, используем другой столбец или создаем фиктивный
    colors = None

# Размер сетки для каждой картинки
grid_size = 4
total_features = len(numeric_columns)
num_plots = (total_features + grid_size - 1) // grid_size

# Создаем картинки по 4x4 с цветами по RainTomorrow
for plot_num in range(num_plots):
    start_idx = plot_num * grid_size
    end_idx = min((plot_num + 1) * grid_size, total_features)
    current_features = numeric_columns[start_idx:end_idx]
    
    print(f"\nКартинка {plot_num + 1}: признаки {start_idx + 1}-{end_idx}")
    
    # Создаем сетку 4x4 для текущей картинки
    fig, axes = plt.subplots(grid_size, grid_size, figsize=(20, 20))
    
    # Если axes не массив 4x4, преобразуем
    if grid_size == 1:
        axes = np.array([[axes]])
    elif axes.ndim == 1:
        axes = axes.reshape(1, -1)
    
    for i, col1 in enumerate(current_features):
        for j, col2 in enumerate(current_features):
            if i < len(current_features) and j < len(current_features):
                if i != j:
                    # Scatter plot с цветами по RainTomorrow
                    if 'RainTomorrow' in df.columns:
                        # Разделяем данные на Yes и No
                        mask_yes = df['RainTomorrow'] == 'Yes'
                        mask_no = df['RainTomorrow'] == 'No'
                        
                        # Рисуем точки для No (синие)
                        axes[i, j].scatter(df.loc[mask_no, col2], df.loc[mask_no, col1], 
                                          alpha=0.5, s=1, color='blue', label='No Rain' if i==0 and j==1 else "")
                        
                        # Рисуем точки для Yes (красные)
                        axes[i, j].scatter(df.loc[mask_yes, col2], df.loc[mask_yes, col1], 
                                          alpha=0.5, s=1, color='red', label='Rain' if i==0 and j==1 else "")
                        
                        # Добавляем легенду только на первый график
                        if i == 0 and j == 1:
                            axes[i, j].legend(fontsize=8)
                    else:
                        # Если нет RainTomorrow, используем обычный scatter
                        axes[i, j].scatter(df[col2], df[col1], alpha=0.5, s=1, color='blue')
                    
                    axes[i, j].set_xlabel(col2, fontsize=8)
                    axes[i, j].set_ylabel(col1, fontsize=8)
                    axes[i, j].tick_params(axis='both', which='major', labelsize=6)
                else:
                    # На диагонали - гистограммы с разделением по RainTomorrow
                    if 'RainTomorrow' in df.columns:
                        mask_yes = df['RainTomorrow'] == 'Yes'
                        mask_no = df['RainTomorrow'] == 'No'
                        
                        axes[i, j].hist(df.loc[mask_no, col1], bins=20, alpha=0.7, 
                                       color='blue', label='No Rain', density=True)
                        axes[i, j].hist(df.loc[mask_yes, col1], bins=20, alpha=0.7, 
                                       color='red', label='Rain', density=True)
                        if i == 0 and j == 0:
                            axes[i, j].legend(fontsize=8)
                    else:
                        axes[i, j].hist(df[col1], bins=30, alpha=0.7, color='skyblue')
                    
                    axes[i, j].set_title(col1, fontsize=10)
                    axes[i, j].tick_params(axis='both', which='major', labelsize=6)
            else:
                # Скрываем пустые subplots
                axes[i, j].set_visible(False)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Автоматически определяем числовые признаки
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()

print(f"Найдено числовых признаков: {len(numeric_columns)}")
print(f"Числовые признаки: {numeric_columns}")

if len(numeric_columns) > 0:
    # Вычисляем матрицу корреляций только для числовых признаков
    correlation_matrix = df[numeric_columns].corr()
    
    # Визуализируем тепловую карту корреляций
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
                fmt='.2f', linewidths=0.5, cbar_kws={'shrink': .8})
    plt.title(f'Матрица корреляций числовых признаков\n(всего {len(numeric_columns)} признаков)', 
              fontsize=16, pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
else:
    print("В данных не найдено числовых признаков!")
```
